```python
import tkinter as tk
from tkinter import messagebox as msgbox

# Handling button click event
def button_click():
    # Show an information message box
    msgbox.showinfo('Information', 'Welcome to cos102 GUI app! \n')

    # Ask for user confirmation
    result = msgbox.askyesno('Confirmation', 'Do you want to continue?')

# Create the main window
root = tk.Tk()
root.title('Home Page')
root.geometry('300x100')

# Add a label widget
label = tk.Label(root, text="Hello Friend \n")
label.pack()

# Add a button widget
button = tk.Button(root, text="Click me", command=button_click)
button.pack()

# Styling the button widget
button.config(fg='green', bg='yellow')

# Start the event loop
root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox
import csv

# Load the employee data from CSV
def load_employees():
    employees = []
    with open("GIG-logistics.csv", mode="r") as file:
        reader = csv.DictReader(file)
        for row in reader:
            employees.append({
                "name": row["name"].strip().lower(),
                "department": row["department"].strip().lower()
            })
    return employees

# Function to check employee
def check_employee():
    name_input = name_entry.get().strip().lower()
    dept_input = dept_entry.get().strip().lower()

    found = False
    same_dept_members = []

    for emp in employees:
        if emp["name"] == name_input and emp["department"] == dept_input:
            found = True
        if emp["department"] == dept_input:
            same_dept_members.append(emp["name"])

    if found:
        welcome_msg = f"Welcome {name_input.title()}!\nOther members in {dept_input.title()} department:\n"
        others = [n.title() for n in same_dept_members if n != name_input]
        welcome_msg += "\n".join(others) if others else "You're the only one here!"
        messagebox.showinfo("Employee Found", welcome_msg)
    else:
        messagebox.showinfo("Not Found", "Sorry, this employee does not exist in the system.")

# GUI Setup
root = tk.Tk()
root.title("Employee Checker")
root.geometry("400x250")

tk.Label(root, text="Enter Employee Name:").pack(pady=5)
name_entry = tk.Entry(root, width=40)
name_entry.pack()

tk.Label(root, text="Enter Department:").pack(pady=5)
dept_entry = tk.Entry(root, width=40)
dept_entry.pack()

tk.Button(root, text="Check", command=check_employee).pack(pady=20)

# Load employees when program starts
employees = load_employees()

root.mainloop()
```
